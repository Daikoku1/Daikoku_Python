```python
from datetime import date
def txt2csv (source_folder) :
    txt_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    result = []
    for txt in txt_files :
        if txt[-3:] != 'pdf':
            time = txt.split('(')
            time = time[2][:-5]
        
            txt_filepath = source_folder + txt    
            with open(txt_filepath, 'rb') as f:
                text = f.read()

            new_dict = {
                'time' : time,
                'text' : text
            }
            result.append(new_dict)

    df = pd.DataFrame(result)                
    temp_name = f"금통위의사록(텍스트파일).xlsx"
    txt_name = os.path.join(source_folder, temp_name)
    excel_writer = pd.ExcelWriter(txt_name, engine = 'xlsxwriter')

    df.to_excel(excel_writer)

    excel_writer.save()
```

```python
import re
for i in range(len(data)):
    data.loc[i, 'text'] = re.sub('- \d+ -', '', data.loc[i, 'text'])
    data.loc[i, 'text'] = re.sub('―|｢|｣|-|[․/→←+]', '', data.loc[i, 'text'])
data.head(2)
```

#### 섹션분리용 함수  
 - tidy_sentences : 강사님 제공 함수에서 1줄코드를 수정함
 - txt_split(data) : 별도 제작

```python
def tidy_sentences(section):
    sentence_enders = re.compile(r'((?<=[함음됨임봄짐움])+[ \s\n\.;\(]+|(?<=다)\.)\s+')
    splits = list((m.start(), m.end()) for m in re.finditer(sentence_enders, section))
    starts = [0] + [i[1] for i in splits]
    ends = [i[0] for i in splits]
    sentences = [section[start:end] for start, end in zip(starts[:-1], ends)]
    for i, s in enumerate(sentences):
        sentences[i] = (s.replace('\n', ' ').replace(' ', ' ')) + '.'

    text = '\n'.join(sentences) if len(sentences) > 0 else ''
    return sentences, text
```

```python
def txt_split(data):
    result_list = []
    for i in range(len(data)):
        time = data.loc[i, 'date']
        text = data.loc[i, 'text']

        p1 = re.search('(.?외환.?국제금융\s?동향.?과 관련하여.*|\(나\) 외환.국제금융\s?(및 금융시장)?\s?동향)\n?\s*(일부 위원은|대부분의 위원들은)|｢?.?외환.?국제금융\s?동향.?｣?과 관련하여.*' , text)
        p2 = re.search('(.?금융시장\s?동향.?과 관련하여,?|\(다\) 금융시장\s?동향)\n?\s*일부 위원은|｢?.?금융시장\s?동향.?｣?과 관련하여,?', text)
        p3 = re.search('((\((다|라)\) )?.?통화정책\s?방향.?에 관한 토론,?|이상과 같은 의견\s?교환을 바탕으로.*통화정책\s?방향.*에.*토론.*)\n?', text)
        
        s1 = p1.start() if p1 else -1
        s2 = p2.start() if p2 else -1
        s3 = p3.start() if p2 else -1
        
        section_2 = text[s1:s2] if s1 >= 0 or s2 >= 0 else ''
        section_3 = text[s2:s3] if s2 >= 0 or s3 >= 0 else ''
        
        sentences2, text2 = tidy_sentences(section_2)
        sentences3, text3 = tidy_sentences(section_3)
        new_dict = {
            'date' : time,
            's2' : text2,
            's3' : text3
        }
        result_list.append(new_dict)
        
    df = pd.DataFrame(result_list)
    return df
```

```python
from ekonlpy.sentiment import MPCK

def text2ngram(text):
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    
    # 불용어 처리
    stop_words = ['NR','VX','VCP','VCN','MM','MAJ','IC','JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JX','JC','EP','EF','EC','ETN','ETM','XPN','XSN','XSV','XSA','XR','SF','SE','SSO','SSC','SC','SY','SL','SH','SN']
    new_tokens = []
    for token in bef_tokens:
        t = token.split('/')[1]
        if t not in stop_words:
            stop_words.append(token)
    return bef_tokens + ngrams
```

```python
for i in range(len(df_t)):
    txt1 = df_t.loc[i, 's2']
    txt2 = df_t.loc[i, 's3']
    
    tokens = text2ngram(txt1) + text2ngram(txt2)
    df_t.loc[i, 'ngrams'] = ','.join(tokens)
    
    if i%10 == 0:
        print(i)
```

# ngram 카운트하기

```python
def count_data(df):
    df_up = pd.DataFrame(df[df['up_down'] == '상승'])
    df_up = df_up.reset_index(drop=True)
    df_down = pd.DataFrame(df[df['up_down'] == '하락'])
    df_down = df_down.reset_index(drop=True)
    
    up_list = []
    down_list = []
    
    # up/down 카운트한 데이터 각각 만들기
    for i in range(len(df_up)):
        up_list.extend(df_up.loc[i, 'ngrams'].split(','))
    dfup = pd.DataFrame(up_list, columns = ['ngram'])
    dataup = dfup['ngram'].value_counts()
    up_df = pd.DataFrame(dataup)
    # up_df 만들 때 해당 데이터 이름으로 컬럼명이 입력됨
    # 별도의 컬럼명을 입력하면 데이터가 다 지워짐
    # 아래 코드로 별도 작성 필요
    up_df.columns = ['up_counts']
    
    for i in range(len(df_down)):
        down_list.extend(df_down.loc[i, 'ngrams'].split(','))
    dfdown = pd.DataFrame(down_list, columns = ['ngram'])
    datadown = dfdown['ngram'].value_counts()
    down_df = pd.DataFrame(datadown)
    down_df.columns = ['down_counts']
    
    idx_list = up_list + down_list
    # 중복된 ngram제거
    idx_list = list(set(idx_list))
    
    # 만들어진 ngram를 인덱스로 하는 데이터프레임 생성
    updown_data=pd.DataFrame(index=idx_list)
    updown_data = pd.concat([updown_data, up_df, down_df],axis=1,sort=False)
    
    # 비어있는 열 삭제
    up_down = updown_data[updown_data.index != '']
    # 확률 era를 막기 위해 빈값에 1 입력
    up_down = up_down.fillna(1)
    return up_down
```
