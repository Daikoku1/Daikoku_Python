```python
import os
b_path = 'C:\\Users\\user\\Desktop\\data\\geopd'
boundary_path = os.path.join(b_path, 'Plate_Boundaries/Plate_Boundaries/Plate_Boundaries.shp')
plate_boundaries = gpd.read_file(boundary_path)
plate_boundaries.head()
```

```python
# 좌표 축 변환
plate_boundaries.apply(lambda x: [(b,a) for (a,b) in list(x.geometry.coords)], axis = 1)
```

```python
plate_boundaries['coordinates'] = plate_boundaries.apply(lambda x: [(b,a) for (a,b) in list(x.geometry.coords)], axis='columns')
plate_boundaries.drop('geometry', axis=1, inplace=True)

plate_boundaries.head()
```
