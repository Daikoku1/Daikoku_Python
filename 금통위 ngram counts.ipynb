# 함수화(금통위 - ngram)

```python
import pandas as pd
def count_data(df):
    df_up = pd.DataFrame(df[df['call'] == '상승'])
    df_up = df_up.reset_index(drop=True)
    df_down = pd.DataFrame(df[df['call'] == '하락'])
    df_down = df_down.reset_index(drop=True)
    
    up_list = []
    down_list = []
    
    for i in range(len(df_up)):
        if df_up.loc[i, 'final_t'] != 0:
            up_list.extend(df_up.loc[i, 'final_t'].split(','))
    dfup = pd.DataFrame(up_list, columns = ['ngram'])
    dataup = dfup['ngram'].value_counts()
    up_df = pd.DataFrame(dataup)
    up_df.columns = ['up_counts']
    
    for i in range(len(df_down)):
        if df_down.loc[i, 'final_t'] != 0:
            down_list.extend(df_down.loc[i, 'final_t'].split(','))
    dfdown = pd.DataFrame(down_list, columns = ['ngram'])
    datadown = dfdown['ngram'].value_counts()
    down_df = pd.DataFrame(datadown)
    down_df.columns = ['down_counts']
    
    return up_df, down_df
```

```python
a , b = count_data(data1)
```

```python
a.to_excel('의사록상승ngram데이터.xlsx')
```

```python
b.to_excel('의사록하락ngram데이터.xlsx')
```

```python
updown_data=pd.DataFrame(index=idx_list)
```

```python
updown_data = pd.concat([updown_data, a, b],axis=1,sort=False)
updown_data.head()
```
