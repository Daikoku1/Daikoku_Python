```python
import pandas as pd
```

```python
up_data = pd.read_excel('의사록상승ngram데이터.xlsx')
down_data = pd.read_excel('의사록하락ngram데이터.xlsx')
```

```python
up_data.columns = ['ngram', 'up_counts']
up_data=up_data.set_index('ngram')
up_data=up_data[up_data['up_counts']>=15]
```

```python
down_data.columns = ['ngram', 'down_counts']
down_data=down_data.set_index('ngram')
down_data=down_data[down_data['down_counts']>=15]
```

```python
lidx=[]
lidx.extend(up_data.index)
lidx.extend(down_data.index)
lidx=list(set(lidx))
```

```python
updown_data=pd.DataFrame(index=lidx)
```

```python
updown_data = pd.concat([updown_data,up_data,down_data],axis=1,sort=False)
```

```python
updown_data = updown_data.fillna(1)
updown_data
```

```python
updown_data.loc[:,'nbc']=0
```

```python
tot_pos=sum(updown_data['up_counts'])
tot_neg=sum(updown_data['down_counts'])
```

```python
for idx in updown_data.index:
    updown_data.loc[idx,'nbc']=(updown_data.loc[idx,'up_counts']/tot_pos)/(updown_data.loc[idx,'down_counts']/tot_neg)
```

```python
updown_data=updown_data[(updown_data['nbc']<0.73) | (updown_data['nbc']>1.3)]
updown_data
```

```python
for idx in updown_data.index:
    if len(idx.split(';')) > 1 :
        print(idx,updown_data.loc[idx,'nbc'])
```
