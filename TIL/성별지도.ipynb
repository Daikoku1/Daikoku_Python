```python
import geopandas as gpd
```

```python
data_path = 'C:\\Users\\user\\Desktop\\30.수원시_성연령별_거주인구격자(250M X 250M).geojson'
df = gpd.read_file(data_path)
print(df.crs)
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df = df.to_crs({'init':'epsg:4326'})
print(df.crs)
df.head()
```

```python
def change(x):
    return sum(x[i] for i in ['w_20g_pop', 'w_30g_pop', 'w_40g_pop', 'w_50g_pop', \
        'w_60g_pop', 'w_70g_pop', 'w_80g_pop', 'w_90g_pop', 'w_100g_pop'])

df['여성인구수'] = df.apply(change, axis = 1)
df.head()
```

```python
df.columns
```

```python
def change(x):
    return sum(x[i] for i in ['m_20g_pop', 'm_30g_pop', 'm_40g_pop', 'm_50g_pop', \
        'm_60g_pop', 'm_70g_pop', 'm_80g_pop', 'm_90g_pop', 'm_100_pop'])

df['남성인구수'] = df.apply(change, axis = 1)
df.head()
```

```python
df = df[['gid', '남성인구수', '여성인구수', 'geometry']]
df.head()
```

```python
print(len(df))
df['gid'].nunique()
```

```python
import folium
from folium import Circle
```

```python
df.loc[0, 'geometry'].centroid.x
```

```python
# Create a base map
m_1 = folium.Map(location=[37.262222, 127.028889], tiles='cartodbpositron', zoom_start=13)
def color_producer(m, w):
    if m < 0.9 * w: return 'red'
    elif m > 1.1 * w : return 'blue'
    else : return 'green'

# Add a bubble map to the base map
for i in range(0,len(df)):
    Circle(
        location=[df['geometry'].centroid.y.iloc[i], df['geometry'].centroid.x.iloc[i]],
        radius=20,
        color=color_producer(df.loc[i, '남성인구수'], df.loc[i, '여성인구수'])).add_to(m_1)

# Display the map
m_1
```
