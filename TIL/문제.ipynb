```python
import pandas as pd
df = pd.read_excel('./data/Quiz.xlsx')
df.head()
```

```python
def check_examples5(x):
    if x == '@': return ''
    else: return x
df['examples5'] = df['examples5'].apply(check_examples5)
df.head()
```

```python
def check_type(x, y):
    if x == True : return 'video'
    if y == True : return 'image'
    return 'text'
df['type'] = df.apply(lambda x : check_type(x['video'], x['image']), axis = 1)
df.head()
```

```python
import json

file_path = "./quiz.json"

data = {}
data['QUIZ'] = []
```

```python
df.columns
```

```python
for num, problem, example1, example2, example3, example4, example5, explanation, answer, video, image, type in df.values:
    data['QUIZ'].append({
        'problem' : problem,
        'examples1' : example1,
        'examples2' : example2,
        'examples3' : example3,
        'examples4' : example4,
        'examples5' : example5,
        'explanation' : explanation,
        'answer' : answer,
        'type' : type
    })
```

```python
len(data['QUIZ'])
```

```python
with open(file_path, 'w') as outfile:
    json.dump(data, outfile)
```
