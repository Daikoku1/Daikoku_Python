```python
import pandas as pd
```

```python
data = pd.read_excel('금통위 문자열.xlsx', usecols = [1, 2, 3])
data.head()
```

```python
data = data.fillna('')
data.head()
```

```python
data.loc[1, 's2']
```

```python
data['text'] = 0
for i in range(len(data)):
    data.loc[i, 'text'] = data.loc[i, 's2'] + data.loc[i, 's3']
data.head()
```

```python
text_file = data[data['text'] != '']
text_file.head()
```

```python
text_file.to_csv('의사록텍스트파일.csv')
```
