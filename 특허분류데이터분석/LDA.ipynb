```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data\활용특허데이터.xlsx')
df.head(1)
```

```python
from konlpy.tag import Komoran
tokenizer = Komoran()
```

```python
import re
df['desc'] = df['독립항'].apply(lambda x : re.sub('(\[\w+\])', '', x))
df['desc'] = df['desc'].apply(lambda x : re.sub('\n', ' ', x))
df['desc'] = df['desc'].apply(lambda x : re.sub('\W', ' ', x))
df['desc'] = df['desc'].apply(lambda x : re.sub('[0-9]+', '', x))

print(df['desc'])
```

```python
def chage(x):
    if x.strip() == '': return None
    else: return x
df['desc'] = df['desc'].apply(chage)
df.dropna(inplace = True, subset = ['desc'])
df['desc']
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df['desc'] = df['desc'].apply(lambda x : tokenizer.nouns(x))
```

```python
df['desc']
```

```python
stop_word = [
    '소폭', '타스', '초단', '조보', '포함', '형성',
    '동리', '테이', '며칠', '터보차저', '중립', '모수',
    '퍼즐', '김로', '강스', '모트', '섹터', '기반','연도', 
    '메인 주', '프리젠테이션', '검토', '쏘파', '랜스', 
    '카나', '카페', '귀의', '에미', '우선', '기계로', 
    '메일', '윤성인', '레코', '업종', '마블', '균일', '족하', 
    '아니', '차차', '정점', '브레이', '비고', '오닐', '경향',
    '엑스', '패커', '상기', '단계', '방법'
]
def remove_stopword(x):
    arr = []
    for i in x:
        if len(i) == 1: 
            if i not in ['벼', '뜸', '약', '팥', '칡', '책', '옷', '숯', '손', '논', '짚', '닭', '땀', '당', '균', '밤', '뱀', '밥', '귤', '떡', '뼈', '햄']:
                continue
        if i in stop_word: continue
        arr.append(i)
    return arr

df['desc'] = df['desc'].apply(remove_stopword)
```

```python
df['desc']
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(df['desc'])
corpus = [dictionary.doc2bow(text) for text in df['desc']]
```

```python
import gensim
NUM_TOPICS = 12 #12개의 토픽
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
vis = gensimvis.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
pyLDAvis.save_html(vis, './result/lda.html')
```
